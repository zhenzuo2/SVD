```python
k= input('Enter how many eigenvalues do you want:')
k=int(k)
from PIL import Image
from numpy import*
import numpy
import scipy
temp=asarray(Image.open('grass_PNG4930.png'))
temp.shape
print(temp)
red=numpy.zeros((temp.shape[0],temp.shape[1]))
green=numpy.zeros((temp.shape[0],temp.shape[1]))
blue=numpy.zeros((temp.shape[0],temp.shape[1]))
one=numpy.zeros((temp.shape[0],temp.shape[1]))
for i in range(temp.shape[0]):
    for j in range(temp.shape[1]):
        red[i][j]=temp[i][j][0]
        green[i][j]=temp[i][j][1]
        blue[i][j]=temp[i][j][2]
        one[i][j]=temp[i][j][3]
```

```python
from getdecom import getdecom
from terms import useterm
newred=useterm(k,red)
newgreen=useterm(k,green)
newblue=useterm(k,blue)
```

```python
newtemp=numpy.zeros((temp.shape[0],temp.shape[1],temp.shape[2]),dtype=numpy.uint8)
```

```python
for i in range(temp.shape[0]):
    for j in range(temp.shape[1]):
        newtemp[i][j][0]=newred[i][j]
        newtemp[i][j][1]=newgreen[i][j]
        newtemp[i][j][2]=newblue[i][j]
        newtemp[i][j][3]=one[i][j]
```

```python
img=Image.fromarray(newtemp,'RGBA')
img.save("{}.png".format(k))
```
